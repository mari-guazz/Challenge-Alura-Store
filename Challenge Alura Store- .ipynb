### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
#Calculo do faturamento por loja
soma_preco_loja1 = loja['Preço'].sum()
soma_preco_loja2 = loja2['Preço'].sum()
soma_preco_loja3 = loja3['Preço'].sum()
soma_preco_loja4 = loja4['Preço'].sum()

print(f"Faturamento Loja 1: R$ {soma_preco_loja1:,.2f}")
print(f"Faturamento Loja 2: R$ {soma_preco_loja2:,.2f}")
print(f"Faturamento Loja 3: R$ {soma_preco_loja3:,.2f}")
print(f"Faturamento Loja 4: R$ {soma_preco_loja4:,.2f}")
```

# 2. Vendas por Categoria


```python
#Venda de produtos por categoria
print("Vendas por Categoria - Loja 1:")
display(loja['Categoria do Produto'].value_counts())

print("\nVendas por Categoria - Loja 2:")
display(loja2['Categoria do Produto'].value_counts())

print("\nVendas por Categoria - Loja 3:")
display(loja3['Categoria do Produto'].value_counts())

print("\nVendas por Categoria - Loja 4:")
display(loja4['Categoria do Produto'].value_counts())
```

# 3. Média de Avaliação das Lojas

```python
# Calcular a média da avaliação da compra para cada loja
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de Avaliação da Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de Avaliação da Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de Avaliação da Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de Avaliação da Loja 4: {media_avaliacao_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
# Separar itens mais e menos vendidos de cada loja

print("Produtos Mais e Menos Vendidos - Loja 1:")
produto_counts_loja1 = loja['Produto'].value_counts()
print("\nMais Vendidos:")
display(produto_counts_loja1.head())
print("\nMenos Vendidos:")
display(produto_counts_loja1.tail())

print("\nProdutos Mais e Menos Vendidos - Loja 2:")
produto_counts_loja2 = loja2['Produto'].value_counts()
print("\nMais Vendidos:")
display(produto_counts_loja2.head())
print("\nMenos Vendidos:")
display(produto_counts_loja2.tail())

print("\nProdutos Mais e Menos Vendidos - Loja 3:")
produto_counts_loja3 = loja3['Produto'].value_counts()
print("\nMais Vendidos:")
display(produto_counts_loja3.head())
print("\nMenos Vendidos:")
display(produto_counts_loja3.tail())

print("\nProdutos Mais e Menos Vendidos - Loja 4:")
produto_counts_loja4 = loja4['Produto'].value_counts()
print("\nMais Vendidos:")
display(produto_counts_loja4.head())
print("\nMenos Vendidos:")
display(produto_counts_loja4.tail())
```

# 5. Frete Médio por Loja


```python
# Média do frete de cada loja
media_frete_loja1 = loja['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

print(f"Média do Frete da Loja 1: R$ {media_frete_loja1:,.2f}")
print(f"Média do Frete da Loja 2: R$ {media_frete_loja2:,.2f}")
print(f"Média do Frete da Loja 3: R$ {media_frete_loja3:,.2f}")
print(f"Média do Frete da Loja 4: R$ {media_frete_loja4:,.2f}")
```

# Gráficos

```python
import matplotlib.pyplot as plt

soma_precos = {
    'Loja 1': soma_preco_loja1,
    'Loja 2': soma_preco_loja2,
    'Loja 3': soma_preco_loja3,
    'Loja 4': soma_preco_loja4
}

lojas = list(soma_precos.keys())
faturamento = list(soma_precos.values())


plt.figure(figsize=(8, 8))
plt.pie(faturamento, labels=lojas, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.title('Proporção do Faturamento Total por Loja')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

media_avaliacoes = {
    'Loja 1': media_avaliacao_loja1,
    'Loja 2': media_avaliacao_loja2,
    'Loja 3': media_avaliacao_loja3,
    'Loja 4': media_avaliacao_loja4
}

lojas = list(media_avaliacoes.keys())
avaliacoes_medias = list(media_avaliacoes.values())

plt.figure(figsize=(8, 6))
plt.barh(lojas, avaliacoes_medias, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.title('Média de Avaliação por Loja')
plt.xlabel('Média de Avaliação')
plt.ylabel('Loja')
plt.xlim(0, 5)
plt.grid(axis='x')
plt.show()
```

# Relatório de Análise das Lojas da Alura Store

Este relatório tem como finalidade apresentar uma análise exploratória dos dados de vendas, avaliações e logística das quatro lojas da Alura Store. A partir dessa análise, será feita uma recomendação fundamentada sobre qual loja o Sr. João deve considerar vender, com base em desempenho financeiro, percepção dos clientes e eficiência operacional.

**Faturamento total**
Após consolidar os dados das quatro unidades, o faturamento total obtido foi de R$ 2.168.122,54, considerando o valor unitário dos produtos multiplicado pela quantidade vendida.

**Vendas por Categoria**
A análise por categoria revela que os produtos mais vendidos pertencem às seguintes categorias:



*   Eletrodomésticos: 2.694 unidades
*   Móveis: 1.908 unidades
*   Informática: 1.499 unidades

A categoria automotivo apresentou o menor volume de vendas, com apenas 240 unidades comercializadas.

**Avaliação Média por Loja**
As avaliações dos consumidores indicam a seguinte média por loja:

*   Loja 1: 3,98
*   Loja 2: 4,02
*   Loja 3: 4,11
*   Loja 4: 3,99

A Loja 3 se destaca com a melhor avaliação média (4,11), demonstrando maior satisfação do cliente e melhor reputação.

**Produtos em Destaque**

*   Produto mais vendido: Refrigerador (369 unidades)
*   Produto menos vendido: Jogo de Chave Allen (21 unidades)


Os dados reforçam que eletrodomésticos são responsáveis por uma fatia relevante das vendas totais, tanto em volume quanto em valor.

**Frete Médio por Loja**
A média de frete cobrada por loja foi:

*   Loja 1: R$ 60,39

*   Loja 2: R$ 54,21

*   Loja 3: R$ 72,55

*   Loja 4: R$ 58,40

Apesar de ter o maior custo de frete, a Loja 3 compensa com desempenho superior em avaliação e vendas.

**Conclusão**

Com base nos critérios de análise, a loja menos eficiente é a Loja 4, por apresentar: Avaliação mediana dos clientes (3,99), abaixo da média das demais, faturamento inferior ao da Loja 1 e menor destaque que a Loja 3, frete alto (R$ 58,40) e nenhuma liderança em categoria de produto, faturamento ou satisfação.
Dessa forma, recomenda-se que o Sr. João deve vender a Loja 4, por ser a unidade com menor eficiência operacional e comercial no conjunto de critérios avaliados.
